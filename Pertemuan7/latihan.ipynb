```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python



dataA = {'a':[1,2],
    'b':[3,4],
    'c':[5,6],
    'd':[7,8],
    'e':[9,10],
    'f':[11,12]}
df = pd.DataFrame(data = dataA)
df
```

```python
df.groupby(['a']).sum().plot(kind='pie', y='b', autopct='%1.0f%%')
```

```python
df.groupby('a')['b'].nunique().plot(kind='bar')
plt.show()
```

```python
df.plot(x='a', y='b', kind='bar')
```
