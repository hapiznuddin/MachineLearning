```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
product_names = ['Kopiko', 'Torabika', 'Beng Beng', 'Choki-Choki', 'Teh Pucuk', 'Energen']
product_stock = [52, 39, 78, 18, 78, 29]

df_names = pd.DataFrame(product_names, columns =['names'])
df_stock = pd.DataFrame(product_stock, columns =['stock'])

df = df_names.assign(stock = df_stock)

print(df)
```

```python
ax = plt.gca()
df.plot(kind='line', x='names', y='stock', color='blue', ax=ax, marker='o')
plt.show()
```

```python
names = ('Kopiko', 'Torabika', 'Beng Beng', 'Choki-Choki', 'Teh Pucuk', 'Energen')
stock = (52, 39, 78, 18, 78, 29)
explode = (0, 0, 0, 0.1, 0, 0)

plt.pie(stock, labels = names, autopct = '%1.1f%%', explode = explode)
plt.show()
```
